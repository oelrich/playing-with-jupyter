# The R kernel

```r
library(tidyr)
library(dplyr)
library(feather)

gun_toters <- feather::read_feather("gun_toters.feather")
gun_toters %>% head()
```

```r
gun_toters %>% group_by(Year) %>% summarize(mean_deaths = mean(Deaths), mean_guns = mean(Guns))
```

```r
library(ggplot2)
ggplot(gun_toters, aes(x=Deaths,y=Guns)) +
    geom_point(aes(colour=GDP))
```
